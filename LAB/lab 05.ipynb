```python
ITEM=["RICE","CHICKEN","MUTTON","COFFEE","PIZZA","BURGER","BISCUIT","BREAD","EGGS","PULSES"]
QUANTITY=[1,1,1,2,1,2,4,2,7,1]
PRICE=[60,250,700,80,200,60,30,20,6,40]
```

```python

import random



# Create and write to a CSV file (as a text file)

with open('Nilanjana_23122024.txt', 'w') as txtfile:
    txtfile.write("Item\t,Quantity\t,UnitPrice\t,TotalCost\t\t\n")

    for i in range(100):
        r = random.randint(0, 9)
        
        quantity = QUANTITY[r]
        
        unitprice = PRICE[r]
    
        total_cost = quantity * unitprice
        txtfile.write(f"{ITEM[r]}\t,{quantity}\t\t\t,{unitprice}\t\t\t,{total_cost}\t\n")



# Read the text file and print results
with open('Nilanjana_23122024.txt', 'r') as txtfile:
    next(txtfile)  # to exclude the row that contains headers


 # Initialize lists to store data

    rows = [line.strip().split(',') for line in txtfile]
    unq_items = set([row[0] for row in rows])

    # Print results for the first 5 rows and last 10 rows

    print("Results for the first 5 rows:")
    for row in rows[:5]:
        print(f"Item: {row[0]}, Quantity: {row[1]}, Total Cost: {row[3]}")

    print("\nResults for the last 10 rows:")
    for row in rows[-5:]:
        print(f"Item: {row[0]}, Quantity: {row[1]}, Total Cost: {row[3]}")


    # Calculate and print total cost for each unique item

    print("\nTotal cost for each unique item:")
    for item in unq_items:
        total_quantity = sum(int(row[1]) for row in rows if row[0] == item)
        total_cost = sum(float(row[3]) for row in rows if row[0] == item)
        print(f"Item Name: {item}, Total Quantity: {total_quantity}, Total Cost: {total_cost}")


    # Calculate and print min and max prices for each item sold
    
    print("\nMinimum and Maximum Prices for Each Item Sold:")
    
    
    min_price = min(unq_items)
    max_price = max(unq_items)
    print(f"Item with Minimum Price: {min_price}, Item with Maximum Price: {max_price}")
        
    
        
       
        
    
        
    
```
